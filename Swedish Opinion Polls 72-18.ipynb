## Swedish Opinion Poll Data 1972-2018
Data from [SCB](https://www.scb.se/), with a gap between 1981 and 1984.

The parties that have been in the Riksdag are reported individually, and in addition "övriga" (others) sums up all other parties people have expressed their sympathies for.

In addition to this, some common groupings for the right and left wing blocks are included. Since these doesn't contain any extra information, I remove them for now.

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
poll_df = pd.read_csv('ME0201B1.csv', engine='python', skiprows=2, encoding='ISO-8859-1', na_values='..')
```

```python
display(poll_df['parti'].unique())
```

```python
poll_df = pd.melt(poll_df, id_vars=['parti'], value_vars=poll_df.loc[:, poll_df.columns != 'parti'])
poll_df['variable'] = pd.to_datetime(poll_df['variable'], format='%YM%m')
poll_df.columns = ['party', 'date', 'support']
poll_df = poll_df.pivot(index='date', columns='party', values='support')
```

```python
poll_df.head()
```

```python
parties_only = poll_df.loc[:, ['V', 'S', 'MP', 'L', 'C', 'M', 'KD', 'NYD', 'SD', 'övriga']]
party_colors = ['#b00000', '#ed1b34', '#80AA4E', '#0069b4', '#39944A', '#019CDB', '#2D338E', 'orange', '#fedf09', 'grey']
parties_only.plot.area(stacked=True, figsize=(20,10), color=party_colors)
```

```python
# Compare now with 90s
compare_20years = parties_only[parties_only.index.isin(['1998-11-01', '2018-11-01'])]
compare_20years.reset_index(inplace=True)
compare_20years = pd.melt(compare_20years, id_vars=['date'], value_vars=compare_20years.loc[:, compare_20years.columns != 'date'])
compare_20years = compare_20years[compare_20years['party'] != 'NYD']
party_colors_exc_NYD = ['#b00000', '#ed1b34', '#80AA4E', '#0069b4', '#39944A', '#019CDB', '#2D338E', '#fedf09', 'grey']
g = sns.catplot(x="party", y="value", col="date", palette=sns.color_palette(party_colors_exc_NYD), data=compare_20years, kind="bar", height=4);
```
